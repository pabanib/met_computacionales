## Iterar en R

Una de las funciones principales para crear un programa es manejar las iteraciones o bucles

### Instrucción for

for (i in secuencia) {
  
  operación 1
  
  ...
  
  operación final
  
}

**Ejemplo** 

Multiplicación de dos vectores 

```r
a <- c(2,6,4)
b <- c(5,7,2)

rdo <- numeric()  #creamos un vector vacío

for (i in 1:length(a)){
  x <- a[i]*b[i]
  rdo[i] <- x
}

rdo
```

```r
a*b
```

instrucción while

while (condición) {

  operación 1

  ...

  operación final 
  
}


**Ejemplo**

Serie de Fibonacci

```r
max <- 10 
i <- 0  #iterador
rdo <- numeric()

while (i < max) {
  if (length(rdo) < 2) {
     rdo[i] <- 1
      }
  else{
    rdo[i] <- rdo[i-1]+rdo[i-2]
  }
  i <- i+1  #importante para no tener un bucle infinito
}

rdo
```

## Funciones

Estructura

nombre_de_función <- function(variable1, variable2,...){
  
    cuerpo

    return(resultado)
  }


Por ejemplo función que devuelve la serie de Fibonacci

```r
Fibonacci <- function(cantidad_digitos){
  i <- 0  #iterador
  rdo <- numeric()

while (i <= cantidad_digitos) {
  if (length(rdo) < 2) {
     rdo[i] <- 1
      }
  else{
    rdo[i] <- rdo[i-1]+rdo[i-2]
  }
  i <- i+1  #importante para no tener un bucle infinito
}

return(rdo)

}
```

```r
Fibonacci(8)
Fibonacci(6)
```
